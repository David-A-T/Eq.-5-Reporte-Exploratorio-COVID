```python
# Importar: Pandas
import pandas as pd
```

```python
# BD: Covid México
cov = pd.read_csv("COVID19MEXICO.csv", index_col="ID_REGISTRO")
```

```python
cov.info()
```

```python
# Filltrar: mantener casos confirmados
values_to_keep = [1, 2, 3]

cov_filtered = cov[cov['CLASIFICACION_FINAL'].isin(values_to_keep)]
```

```python
# Función: obtener coordenadas
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_geocoder")

def obtener_lat_long(direccion):
    location = geolocator.geocode(direccion)
    print(direccion)
    if location:
        print(location.latitude, " | ", location.longitude)
        return location.latitude, location.longitude
    else:
        return None, None
```

```python
# BD: auxiliares | VESTIGIAL
ent = pd.read_excel("201128 Catalogos.xlsx", sheet_name="Catálogo de ENTIDADES")
ent_aux = ent[["CLAVE_ENTIDAD", "ENTIDAD_FEDERATIVA"]]
ent_aux = ent_aux.set_index("CLAVE_ENTIDAD")
ent_dic = ent_aux.to_dict()
```

```python
# BD: auxiliares | VESTIGIAL
ent['EF_LAT'], ent["EF_LON"] = zip(*ent['ENTIDAD_FEDERATIVA'].apply(obtener_lat_long))
```

```python
# BD: auxiliares
mun = pd.read_excel("201128 Catalogos.xlsx", sheet_name="Catálogo MUNICIPIOS")
mun["ENTIDAD"] = mun["CLAVE_ENTIDAD"]
mun.replace({"ENTIDAD": ent_dic["ENTIDAD_FEDERATIVA"]}, inplace=True)
mun["MUNICIPIO"] = mun["MUNICIPIO"] + ", " + mun["ENTIDAD"]
```

```python
# Generar: coordenadas
mun["M_LAT"], mun["M_LON"] = zip(*mun["MUNICIPIO"].apply(obtener_lat_long))
```

```python
# Exportar: bd municipios con coordenadas
mun.to_csv("mun_coord.csv")
```

```python
# BD: auxiliares
data_aux = pd.DataFrame()
#data_aux["ESTADO_RESIDENCIA"] = mun["CLAVE_ENTIDAD"] % VESTIGIAL
data_aux["MUNICIPIO_RESIDENCIA"] = mun["CLAVE_MUNICIPIO"]
data_aux["ESTADO_RESIDENCIA"] = mun["CLAVE_ENTIDAD"]
data_aux["MUN_LAT"] = mun["M_LAT"]
data_aux["MUN_LON"] = mun["M_LON"]
```

```python
# BD: para análisis
data = data_aux.copy()
data["CASOS_CONFIRMADOS"] = None
```

```python
data
```

```python
# Cuantificar: casos confirmados por municipio
for i in range(data.shape[0]):
    print(i)
    casos = len(cov_filtered[(cov_filtered["MUNICIPIO_RES"] == data.iloc[i, 0])])
    data.loc[i, "CASOS_CONFIRMADOS"] = casos
```

```python
# Cuantificar: casos confirmados de indigenas por municipio
data["INDIGENA"] = None
for i in range(data.shape[0]):
    print(i)
    casos = len(cov_filtered[(cov_filtered["INDIGENA"] == 1) & (cov_filtered["MUNICIPIO_RES"] == data.iloc[i, 0])])
    print(casos)
    data.loc[i, "INDIGENA"] = casos
```

```python
# Cuantificar: casos confirmados de migrantes por municipio
data["MIGRANTE"] = None
for i in range(data.shape[0]):
    print(i)
    casos = len(cov_filtered[(cov_filtered["MIGRANTE"] == 1) & (cov_filtered["MUNICIPIO_RES"] == data.iloc[i, 0])])
    print(casos)
    data.loc[i, "MIGRANTE"] = casos
```

```python
# Depuración: valores nulos
data.dropna(inplace=True)
```
